# Import Essential Libraries

```python
import numpy as np 
import pandas as pd 
```

# Import data from Excel file

```python
data_frame  = pd.read_csv('nutrients_csvfile.csv')
data_frame
```

# Replace empty values in the dataset

```python
data_frame = data_frame.replace("t",0)
data_frame = data_frame.replace("t'",0)
data_frame.head()
```

# Perform transformation on data frame  

```python
data_frame = data_frame.replace(",","", regex=True)
data_frame['Fiber']= data_frame['Fiber'].replace("a","", regex=True)
data_frame['Calories'][91]=(8+44)/2
```

```python
data_frame['Grams']=pd.to_numeric(data_frame['Grams'])
data_frame['Calories']=pd.to_numeric(data_frame['Calories'])
data_frame['Protein']=pd.to_numeric(data_frame['Protein'])
data_frame['Fat']=pd.to_numeric(data_frame['Fat'])
data_frame['Sat.Fat']=pd.to_numeric(data_frame['Sat.Fat'])
data_frame['Fiber']=pd.to_numeric(data_frame['Fiber'])
data_frame['Carbs']=pd.to_numeric(data_frame['Carbs'])
```

```python
data_frame.dtypes
```

# Remove missing values in the data set

```python
data_frame.isnull().sum()
```

```python
data_frame.dropna(inplace = True)
```

```python
data_frame
```

```python
data_frame.isnull().sum()
```

```python
data_frame.describe()
```

# Produce numerical representation of the categorical data in "Category" column and merge data frames

```python
categorical_data_frame = pd.get_dummies(data_frame['Category'])
categorical_data_frame
```

```python
data_frame = pd.concat([data_frame,categorical_data_frame],axis = 1)
```

```python
data_frame.head()
```

```python
data_frame.drop('Category',axis =1 ,inplace = True)
```

```python
data_frame.head()
```

# Generate pivot table of sumarized data_frame data

```python
recommended_data_frame = data_frame.pivot_table(columns='Food',values=['Calories','Fat','Protein','Carbs','Fiber','Sat.Fat'])
recommended_data_frame
```

```python
recommended_data_frame1 = recommended_data_frame['Fortified milk']
```

```python
similar = recommended_data_frame.corrwith(recommended_data_frame1)
```
